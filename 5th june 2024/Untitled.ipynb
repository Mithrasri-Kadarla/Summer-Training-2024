```python
class Node:
    def __init__(self, u):
        self.data = u
        self.next = None

class SLL:
    def __init__(self):
        self.head = None

    def display(self):
        t = self.head
        while t is not None:
            print(t.data, end=" -> ")
            t = t.next
        print("None")

    def add_front(self, x):
        t = Node(x)
        t.next = self.head
        self.head = t

    def add_back(self, x):
        if not self.head:
            self.head = Node(x)
            return
        t = self.head
        while t.next is not None:
            t = t.next
        t.next = Node(x)

    def add_even(self):
        t = self.head
        s = 0
        while t is not None:
            if t.data % 2 == 0:
                s += t.data
            t = t.next
        print(s)

    def search(self, key):
        t = self.head
        while t is not None:
            if t.data == key:
                return True
            t = t.next
        return False

    def middle(self):
        f = self.head
        s = self.head
        while f is not None and f.next is not None:
            s = s.next
            f = f.next.next
        if s:
            print(s.data)

    def check_length(self):
        f = self.head
        s = self.head
        while f is not None and f.next is not None:
            s = s.next
            f = f.next.next
        if f is None:
            print("Even")
        else:
            print("Odd")

    def largest_subsequence(self):
        t = self.head
        c = 1
        m = 1
        while t.next is not None:
            if t.data == t.next.data - 1:
                c += 1
            else:
                if c > m:
                    m = c
                c = 1
            t = t.next
        if c > m:
            m = c
        print(m)

    def all_pairs(self):
        t = self.head
        while t is not None:
            t1 = t.next
            while t1 is not None:
                print(t.data, t1.data)
                t1 = t1.next
            t = t.next

    def bubble_sort(self):
        if not self.head:
            return
        T = self.head
        p = None
        while T.next is not None:
            f = 0
            t = self.head
            while t.next != p:
                if t.data > t.next.data:
                    f = 1
                    t.data, t.next.data = t.next.data, t.data
                t = t.next
            if f == 0:
                break
            p = t
            T = T.next

# Example usage
l1 = SLL()
l1.add_back(10)
l1.add_back(20)
l1.add_back(30)
l1.add_back(40)
l1.add_back(50)
l1.add_back(60)
l1.add_front(7)
l1.add_front(5)
l1.display()

l2 = SLL()
l2.add_back(100)
l2.add_back(200)
l2.add_back(300)
l2.display()

l1.add_even()
l2.add_even()
print(l1.search(30))
l1.middle()
l1.check_length()
l1.largest_subsequence()
l1.all_pairs()

l3 = SLL()
l3.add_back(6)
l3.add_back(7)
l3.add_back(9)
l3.add_back(8)
l3.add_back(4)
l3.add_back(2)
l3.add_back(0)
l3.add_back(1)
l3.display()
l3.largest_subsequence()
l3.bubble_sort()
l3.display()
```
